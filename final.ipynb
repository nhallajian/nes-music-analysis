```python
importlib.reload(patterns4)

# --- Configuration ---
min_pattern_length = 4 # Define min_pattern_length here as it's needed for path construction

# 1. Get all relevant MIDI file paths
# all_midi_files = list(MIDI_ROOT.glob("*.mid")) # Original glob
all_midi_files = list(MIDI_ROOT.glob("*.mid")) # User's specific glob

# 2. Sort the list of paths based on their file size (ascending)
sorted_midi_paths = sorted(all_midi_files, key=lambda path: path.stat().st_size)
print(f"Found {len(sorted_midi_paths)} MIDI files. Sorting by size (low to high).")

# --- Resume Logic ---
resume_index = 0
if sorted_midi_paths: # Only check if there are files
    print("Checking for existing analysis results to resume...")
    last_found_index = -1
    for index, midi_path in enumerate(sorted_midi_paths):
        midi_name = midi_path.stem
        # Construct the expected output directory path for this file
        expected_algorithm_path = POINT_SETS_ROOT / midi_name / f"t_custom_{min_pattern_length}"
        if expected_algorithm_path.exists() and expected_algorithm_path.is_dir():
            # If the directory exists, this file was at least started.
            # Keep track of the latest (largest file size) one found.
            last_found_index = index

    if last_found_index != -1:
        resume_index = last_found_index
        resume_file_name = sorted_midi_paths[resume_index].name
        print(f"Found existing results up to file '{resume_file_name}'. Resuming analysis from this file.")
    else:
        print("No existing results found, starting analysis from the beginning.")
else:
    print("No MIDI files found matching the pattern.")

# --- Main Processing Loop ---
if sorted_midi_paths:
    # Slice the list to start from the resume_index
    files_to_process = sorted_midi_paths[resume_index:]
    print(f"Processing {len(files_to_process)} files starting from index {resume_index}...")

    for i, midi_path in enumerate(files_to_process, start=resume_index):
        iteration_start_time = time.time() # Record start time for this file iteration

        file_size = midi_path.stat().st_size
        print(f"\nProcessing ({i + 1}/{len(sorted_midi_paths)}): {midi_path.name} (Size: {file_size} bytes)")

        midi_name = midi_path.stem
        algorithm_path = POINT_SETS_ROOT / midi_name / f"t_custom_{min_pattern_length}"
        algorithm_path.mkdir(parents=True, exist_ok=True) # Ensure directory exists

        for target_channel in ["p1", "p2", "tr"]:
            min_pattern_occurrences = 2
            output_image_path = algorithm_path / f"{target_channel}.png"

            try:
                 # Call the analysis and visualization function
                 # Ensure patterns4 and its function are correctly imported/defined
                 fig, ax, stats = patterns4.analyze_and_visualize_optimal_patterns(
                     midi_path=midi_path,
                     channel_name=target_channel,
                     min_length=min_pattern_length,
                     min_occurrences=min_pattern_occurrences,
                     save_path=output_image_path
                 )

                 # Save the statistics to a JSON file
                 stats_path = algorithm_path / f"{target_channel}_stats.json"
                 with open(stats_path, "w") as file:
                     # Keep indent for better readability of the JSON output
                     json.dump(stats, file, indent=4)

                 # Optional: Close the figure if it's generated and not needed anymore
                 # if fig:
                 #     import matplotlib.pyplot as plt
                 #     plt.close(fig)

            except Exception as e:
                print(f"  ERROR processing channel {target_channel} for {midi_path.name}: {e}")
                # Decide if you want to continue with other channels or skip the file
                # continue # Skips to the next channel
                # break # Skips to the next file if uncommented

        iteration_end_time = time.time() # Record end time for this file iteration
        iteration_duration = iteration_end_time - iteration_start_time
        # Use i + 1 for the count message
        print(f"({i + 1}/{len(sorted_midi_paths)}) Finished processing {midi_path.name}. Time taken: {iteration_duration:.2f} seconds")

    print("\nFinished processing all MIDI files.")
else:
    print("No MIDI files to process.")
```
